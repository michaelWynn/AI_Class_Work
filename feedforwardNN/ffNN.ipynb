```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
TARGET_VARIABLE = "user_action"
TRAIN_TEST_SPLIT = 0.5
HIDDEN_LAYER_SIZE = 30
raw_data = pd.read_csv("data.csv")
```

```python
raw_data.head(10)
```

```python
raw_data.info()
```

```python
mask = np.random.rand(len(raw_data)) < TRAIN_TEST_SPLIT 
tr_dataset = raw_data[mask]
te_dataset = raw_data[~mask]
```

```python
tr_dataset.head()
```

```python
te_dataset.head()
```

```python
tr_data = np.array(raw_data.drop(TARGET_VARIABLE,axis='columns'))
tr_labels = np.array(raw_data[[TARGET_VARIABLE]])
te_data = np.array(te_dataset.drop(TARGET_VARIABLE,axis='columns'))
te_labels = np.array(te_dataset[[TARGET_VARIABLE]])
```

```python
tr_data
```

```python
tr_labels
```

```python
te_data
```

```python
te_labels
```

```python
## Creates feedforward neural network
ffnn = Sequential()
# Specifies hidden layer size (30), input size (3D vectors as single input), and activation function (sigmoid)
ffnn.add(Dense(HIDDEN_LAYER_SIZE, input_shape=(3,), activation="sigmoid"))
# Specifies output layer with sigmoid activation function
ffnn.add(Dense(1, activation="sigmoid")) 
# Specifies error calculation (MSE), optimizer (stochastic gradient descent), and metrics (accuracy) -> Then compiles
ffnn.compile(loss="mean_squared_error", optimizer="sgd", metrics=['accuracy'])
# Trains feedforward neural network over 150 epochs - verbose prints accuracy & loss after each epoch
ffnn.fit(tr_data, tr_labels, epochs=150, batch_size=2, verbose=1)
```

```python
# Evaluates FFNN with test data and test labels
metrics = ffnn.evaluate(te_data, te_labels, verbose=1) 
print("%s: %.2f%%" % (ffnn.metrics_names[1], metrics[1]*100))
```

```python
# Predicts 'user_action' for new data
new_data = np.array(pd.read_csv("new_data.csv")) 
results = ffnn.predict(new_data)
print(results)
```
